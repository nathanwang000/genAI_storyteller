```python
import os
import numpy as np
from lib.utils import txt2img, img2img, save_img
from skimage.transform import resize
```

```python
img = txt2img('universe, stars', 'ugly, disfigured')
```

```python
n = 100
desired_size = img.shape[0], img.shape[1]
img2 = img[n:-n,n:-n,:]
# resize img back                                                                                      
img2 = resize(img2, desired_size, anti_aliasing=True)
```

```python
# save_img(img, os.path.join(output_dir, f'{i}_resize.png'))
img3 = img2img('universe, stars', 'ugly, disfigured', init_im=img)
```

```python
img4 = img2img('universe, stars', 'ugly, disfigured', init_im=(img2*255).astype(np.uint8))
```

```python
import matplotlib.pyplot as plt
for im in [img, img2, img3, img4]:
    print(im.dtype)
    plt.imshow(im)
    plt.show()
```
